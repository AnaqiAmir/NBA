```python
import pandas as pd
pd.set_option('display.max_rows', 30)
pd.set_option('display.max_columns', 25)
```

```python
# Read data
all_games = pd.read_csv("nba_games.csv")

# Drop unnecessary columns
all_games.drop(['Unnamed: 0', '+/-', 'mp_max', 'mp_max.1', '+/-_max', '+/-_opp',
                'mp_opp.1', 'mp_max_opp', 'mp_max_opp.1', '+/-_max_opp'], axis=1, inplace=True)

# Drop every other row because it contains the same information
all_games = all_games.iloc[::2, :]

# Sort by date
all_games = all_games.sort_values("date")

# Reset index
all_games = all_games.reset_index(drop=True)
```

```python
# Add target variable which is the result of the next game for the home team
def add_target(group):
    group["target"] = group["won"].shift(-1)
    return group

all_games = all_games.groupby("team", group_keys=False).apply(add_target)
```

```python
unwanted_index = all_games[(all_games['target']!=True)&(all_games['target']!=False)].index.tolist()
all_games.drop(unwanted_index,axis=0,inplace=True)
```

```python
all_games.to_csv("nba_games.csv")
```
