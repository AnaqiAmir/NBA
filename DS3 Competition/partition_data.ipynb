```python
# Imports
import pandas as pd

# Inserting game_id column to identift games
df = pd.read_csv("nba_games.csv")
df['game_id'] = df.index
first_column = df.pop('game_id')
df.insert(0, 'game_id', first_column)

# Train set
train = df.sample(frac=0.7)

# Test set
test = df[~df.index.isin(train.index)]

# Solution set
solution = test
solution = solution[['game_id', 'target']]

# Dropping target column in test as participants are required to predict this
test.drop("target",axis=1,inplace=True)

# Creating usage column to separate data into public and private for the Kaggle leaderboards
solution = pd.read_csv("solutions.csv")
solution['Usage'] = ['Public']*3092
solution['Usage'][2000:] = ['Private']*1092

# Creating sandbox submission (simply initialized everything to 1)
sandbox = solution
sandbox.drop('Usage',axis=1,inplace=True)
sandbox['target'] = 1
sandbox.to_csv("sandbox.csv", index=False)

# Saving the files
train.to_csv("train.csv", index=False)
test.to_csv("test.csv", index=False)
solution.to_csv("solutions.csv", index=False)
```
